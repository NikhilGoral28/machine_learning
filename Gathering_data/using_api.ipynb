```python
df = pd.DataFrame()
```

```python
df
```

```python
for i in range(1,533):
    responce = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=4f12fd0505000ec52fba0d34e875c89b&page={i}')
    temp_df = pd.DataFrame(responce.json()['results'], columns=['id','title','vote_average','vote_count','release_date','popularity','overview'])

    df = pd.concat([df,temp_df], ignore_index=True)
```

```python
df.to_csv('top_rated_movies.csv', index=False)
```
