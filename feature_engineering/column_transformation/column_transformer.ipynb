```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['has_covid']), df['has_covid'], test_size=0.2)
```

column transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
#parameter transformers is the for column which are going to transform
#remainder is for the which column dont apply the transformer -- passthough means as it is pass and drop to drop the columns
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3', OneHotEncoder(sparse_output=False  ,drop='first'), ['gender','city'])],
    remainder='passthrough')
```

```python
transformer.fit_transform(X_train)
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.fit_transform(X_test)
```
